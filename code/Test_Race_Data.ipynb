```python
import pandas as pd
import os

# NASCAR race data from GitHub (publicly available)
url = "https://raw.githubusercontent.com/Neil-Paine-1/NASCAR-data/master/nascar-race-data.csv"

print("Downloading NASCAR race data from GitHub...")
print("(This may take a moment - it's a large dataset)")

# Download the CSV directly into pandas
# Note: encoding='latin-1' handles special characters, low_memory=False prevents dtype warnings
df = pd.read_csv(url, encoding='latin-1', low_memory=False)

print(f"\nSuccessfully loaded {len(df):,} race records!")
print(f"\nColumns available:")
print(list(df.columns))
```

```python
# Preview the data
print("First 10 rows:")
print(df.head(10))

print("\n" + "="*50)
print("\nData types:")
print(df.dtypes)

print("\n" + "="*50)
print("\nBasic statistics:")
print(df.describe())
```

```python
# Filter for recent Cup Series races (2023-2024)
# Adjust the year filter based on what's available in the dataset
recent_races = df[df['Date'] >= '2023-01-01'].copy()

print(f"Recent races (2023+): {len(recent_races):,} records")

# Show unique tracks
if 'track' in recent_races.columns:
    print(f"\nTracks in recent data:")
    print(recent_races['track'].value_counts().head(10))

# Show race winners (if available)
if 'winner' in recent_races.columns or 'finish_pos' in recent_races.columns:
    print("\nSample of recent race results:")
    print(recent_races.head(20))
```

```python
import os

# Create directory if it doesn't exist
os.makedirs('data/raw', exist_ok=True)
os.makedirs('data/external', exist_ok=True)

# Save the full dataset for later use
df.to_csv('data/external/nascar_race_data_full.csv', index=False)
print(f"Full dataset saved to data/external/nascar_race_data_full.csv")

# Save a filtered sample for testing
sample = df.tail(1000)  # Last 1000 records
sample.to_csv('data/raw/test_race_results.csv', index=False)
print(f"Test sample saved to data/raw/test_race_results.csv ({len(sample)} rows)")

print("\nâ Race data test complete!")
```
